# 데이터 전처리
데이터 전처리는 원시 데이터를 분석이나 머신러닝 모델에 적합한 형태로 변환하는 과정을 의미합니다. 전처리는 다음과 같은 이유로 중요합니다.

+ __품질 향상__: 잘못된 데이터, 누락된 데이터 또는 잡음이 많은 데이터는 분석 결과의 정확성을 저해할 수 있습니다.
+ __효율적인 모델링__: 더 깔끔하고 구조화된 데이터는 머신러닝 모델 학습에 더 유용하며 더 나은 결과를 도출할 수 있습니다.
+ __분석의 용이성__: 깔끔한 데이터는 데이터 분석가나 연구자가 데이터를 이해하고 분석하는 데 더 용이합니다.

데이터 전처리의 주요 단계와 예시는 다음과 같습니다.

+ __데이터 클렌징__
    - 누락된 값 처리: 예를 들어 평균, 중앙값, 최빈값 등으로 누락된 값을 채움.
    - 이상치 탐지 및 처리: z-점수, IQR 기반 등의 방법으로 이상치를 식별하고 제거하거나 대체.
+ __데이터 변환__
    - 정규화: 데이터의 스케일을 [0, 1] 범위로 조정.
    - 표준화: 데이터의 평균을 0, 표준 편차를 1로 조정.
    - 원-핫 인코딩: 범주형 변수를 0과 1로 구성된 이진 벡터로 변환.
+ __데이터 축소__
    - 차원 축소: PCA(주성분 분석) 등의 기법을 사용하여 변수의 수를 줄임.
    - 특징 선택: 중요한 특징만 선택하여 데이터의 차원을 줄임.
+ __데이터 구조 변경__
    - 재구성: 피벗 테이블이나 데이터 그룹화를 사용하여 데이터 형태 변환.
    - 데이터 결합: 여러 데이터 소스에서 정보를 병합.

**__예시:__**
데이터셋에 "나이"와 "월급"이라는 두 열이 있을 때, "나이" 열에 누락된 값이 있다면 그 값을 해당 열의 평균값으로 대체할 수 있습니다. "월급" 열의 값의 범위가 0에서 100,000 사이라면 이를 [0, 1] 범위로 정규화하여 모든 값을 동일한 스케일로 조정할 수 있습니다.

결론적으로, 데이터 전처리는 높은 품질의 데이터를 확보하여 더 나은 분석 결과를 얻기 위한 필수적인 단계입니다.

"train.csv" 파일의 내용을 확인하겠습니다. 파일의 처음 몇 줄을 보면서 데이터의 구조와 특징을 파악하겠습니다.

데이터의 특성을 확인해보니 다음과 같은 정보들을 파악할 수 있습니다:

> 1. index: 데이터의 인덱스입니다.
> 2. gender: 성별 정보입니다. (F: 여성, M: 남성)
> 3. car: 차량 소유 여부입니다. (Y: 소유, N: 미소유)
> 4. reality: 부동산 소유 여부입니다. (Y: 소유, N: 미소유)
> 5. child_num: 자녀 수입니다.
> 6. income_total: 총 소득입니다.
> 7. income_type: 소득 유형입니다.
> 8. edu_type: 교육 수준 정보입니다.
> 9. family_type: 가족 구조 정보입니다.
> 10. house_type: 주택 유형입니다.
> 11. DAYS_BIRTH: 태어난 지 몇 일이 지났는지의 정보입니다.
> 12. DAYS_EMPLOYED: 취업한 지 몇 일이 지났는지의 정보입니다.
> 13. FLAG_MOBIL: 모바일 전화 소유 여부입니다.
> 14. work_phone: 업무용 전화 소유 여부입니다.
> 15. phone: 전화 소유 여부입니다.
> 16. email: 이메일 소유 여부입니다.
> 17. occyp_type: 직업 유형입니다.
> 18. family_size: 가족 크기입니다.
> 19. begin_month: 얼마나 오래 전에 회사를 시작했는지에 대한 정보입니다.
> 20. credit: 신용 등급입니다.


전처리가 필요한 이유는 다음과 같습니다:

- __결측값 처리__: occyp_type에서 NaN 값을 확인할 수 있습니다. 이와 같은 결측값을 처리해야 합니다.
- __범주형 데이터 인코딩__: gender, car, reality, income_type, edu_type, family_type, house_type, occyp_type 등의 문자열 데이터를 모델 학습에 사용할 수 있도록 숫자로 변환해야 합니다.
- __이상치 처리__: 데이터에 이상치가 있는 경우 이를 처리하거나 제거해야 합니다.
- __새로운 특성 생성__: DAYS_BIRTH나 DAYS_EMPLOYED와 같은 정보를 이용하여 사용자의 나이나 근무 연수와 같은 새로운 특성을 생성할 수 있습니다.
- __데이터 스케일링__: income_total, DAYS_BIRTH, DAYS_EMPLOYED 등의 숫자 데이터의 범위가 다르므로 이를 동일한 범위로 스케일링해야 합니다.
- __정규화__ : 정규화를 수행함으로써, 여러 알고리즘에서 더 나은 성능을 기대할 수 있습니다. 특히, 특성 간의 스케일 차이에 민감한 알고리즘에서는 이러한 정규화가 중요합니다.<br>

전처리 과정은 데이터의 특성과 목적에 따라 달라질 수 있으므로, 무엇을 예측하거나 분석하려는지에 따라 필요한 전처리 단계를 결정해야 합니다. 여기서는 기본적인 전처리 과정만을 설명드렸습니다.

## 결측값 처리
먼저 데이터의 결측값을 확인하고 전처리를 수행해보겠습니다.
결측값 처리는 데이터 전처리 과정에서 중요한 단계 중 하나입니다. 데이터에 결측값이 존재할 경우, 대부분의 머신러닝 알고리즘이 제대로 작동하지 않기 때문에 이러한 결측값을 적절히 처리하는 것이 필요합니다.
### 결측값 처리 방법
1. __제거__: 결측값이 포함된 행 또는 열을 완전히 제거합니다. 이 방법은 데이터 손실이 발생할 수 있습니다.
2. __평균, 중앙값, 최빈값으로 대체__: 수치형 데이터의 경우 평균 또는 중앙값으로, 범주형 데이터의 경우 최빈값으로 결측값을 대체합니다.
3. __보간__: 시계열 데이터에서 주로 사용되며, 앞뒤의 데이터를 기반으로 결측값을 추정합니다.
4. __모델을 사용한 대체__: 다른 변수들을 사용하여 결측값을 가진 변수를 예측하는 모델(예: 회귀, KNN)을 학습시킨 후, 이를 사용하여 결측값을 대체합니다.

### 예시
Python의 pandas 라이브러리를 사용하여 결측값 처리를 진행하는 예시 코드입니다.

````python
# 예제 데이터 생성
data = {
    'Age': [25, 30, 35, None, 40],
    'Salary': [50000, 55000, None, 62000, None]
}
df = pd.DataFrame(data)

# 1. 결측값 제거
df_dropna = df.dropna()

# 2. 평균값으로 대체
df_filled_mean = df.fillna(df.mean())

# 3. 중앙값으로 대체
df_filled_median = df.fillna(df.median())

# 4. 최빈값으로 대체 (주로 범주형 변수에 사용)
# mode()는 Series를 반환하므로 [0]을 사용하여 첫 번째 값을 선택
df_filled_mode = df['Age'].fillna(df['Age'].mode()[0])

print("데이터")
print(df)
print("\n1. 결측값 제거")
print(df_dropna)
print("\n2. 평균값으로 대체")
print(df_filled_mean)
print("\n3. 중앙값으로 대체")
print(df_filled_median)
print("\n4. 최빈값으로 대체 (주로 범주형 변수에 사용)")
print(df_filled_mode)
````

위의 코드에서는 간단한 예제 데이터를 생성하고, 다양한 방법으로 결측값을 처리하는 방법을 보여줍니다. 실제 데이터에 따라 가장 적절한 방법을 선택하여 적용해야 합니다.

## 범주형 변수 인코딩

범주형 변수 인코딩은 머신러닝 모델을 학습하기 전에 수행해야 하는 중요한 전처리 단계 중 하나입니다. 대부분의 머신러닝 알고리즘은 숫자형 데이터만 처리할 수 있기 때문에 범주형 데이터를 적절한 숫자형 데이터로 변환해야 합니다.
### 범주형 변수 인코딩 방법
1. __Label Encoding__: 범주 각각에 고유한 정수를 할당합니다. 주로 순서가 있는 범주형 데이터에 사용됩니다.
2. __One-Hot Encoding__: 범주 각각을 이진 특징으로 변환합니다. 이 방법은 순서가 없는 범주형 데이터에 주로 사용됩니다.

### 예시

````python
# import pandas as pd
from sklearn.preprocessing import LabelEncoder

# 예제 데이터 생성
data = {
    'Age': [25, 30, 35, 40, 45],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
    'Grade': ['A', 'B', 'A', 'C', 'B']
}
df = pd.DataFrame(data)

# 범주형 변수만 선택
categorical_cols = df.select_dtypes(include=['object']).columns

# Label Encoding
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    df[col + '_label'] = le.fit_transform(df[col])
    label_encoders[col] = le

# One-Hot Encoding
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

print(df)
````

위의 코드에서는 먼저 간단한 예제 데이터를 생성하고, df.select_dtypes를 사용하여 범주형 변수만 선택합니다. 그 후, Label Encoding과 One-Hot Encoding을 각각 수행합니다. LabelEncoder와 One-Hot Encoding은 sklearn.preprocessing에서 제공하는 도구입니다. Label Encoding은 각 범주에 고유한 정수를 할당하며, One-Hot Encoding은 가변수 (dummy variable) 방식을 사용하여 범주형 변수에 들어있는 항목의 개수만큼 0 또는 1 값을 가지는 더미변수를 생성해주는 기법입니다.

이제 프로젝트 'train' 데이터에 One-Hot Encoding을 적용합니다.

````python
# object type(범주형)인 변수의 리스트 추출
categorical_cols = train.select_dtypes(include=['object']).columns.tolist()
categorical_cols
````

````python
# 'categorical_cols'에 해당하는 column들을 기반으로 더미변수를 생성합니다.
dum = pd.get_dummies(train, columns=categorical_cols)
# 기존 데이터와 더미변수를 하나로 합칩니다.
train = pd.merge(train, dum, how="left")
train.head()
````

## 이상치 처리
이상치는 다른 관측값들과 크게 다른 값을 가지는 데이터 포인트를 의미합니다. 이상치는 여러 원인(측정 오류, 입력 오류 등)으로 인해 발생할 수 있으며, 이러한 이상치가 포함된 데이터를 사용하여 모델을 학습하면 모델의 성능에 부정적인 영향을 미칠 수 있습니다. 따라서 데이터 전처리 과정에서 이상치를 식별하고 적절하게 처리하는 것이 중요합니다.

### 이상치 탐지 방법
1. __통계적 방법__: 데이터의 평균과 표준편차를 사용하여 이상치를 식별합니다. 예를 들어, 평균에서 3배의 표준편차보다 크거나 작은 값은 이상치로 간주할 수 있습니다.
2. __IQR (사분위수 범위)__: 1사분위(Q1)와 3사분위(Q3)를 계산하고, IQR = Q3 - Q1을 구합니다. Q1 - 1.5IQR 보다 작거나 Q3 + 1.5IQR 보다 큰 값은 이상치로 간주합니다.
3. __시각적 방법__: 박스 플롯, 히스토그램, 산점도 등을 사용하여 데이터를 시각화하고 이상치를 직접 확인합니다.

### 이상치 처리 방법
+ __제거__: 이상치를 포함하는 행을 데이터셋에서 제거합니다.
+ __대체__: 이상치를 다른 값(예: 중앙값)으로 대체합니다.
+ __변환__: 로그 변환과 같은 기법을 사용하여 이상치의 영향을 줄입니다.

### 예시
다음은 pandas와 numpy를 사용하여 IQR을 기반으로 이상치를 탐지하고 제거하는 예시 코드입니다:

````python
# 예제 데이터 생성
data = {
    'Value': [15, 14, 16, 15, 100, 14, 13, 15, 200, 16]
}
df = pd.DataFrame(data)

# IQR을 사용하여 이상치 탐지
Q1 = df['Value'].quantile(0.25)
Q3 = df['Value'].quantile(0.75)
IQR = Q3 - Q1
filter = (df['Value'] >= Q1 - 1.5 * IQR) & (df['Value'] <= Q3 + 1.5 * IQR)
df_filtered = df[filter]

print("Original Data")
print(df)
print("\nData after removing outliers")
print(df_filtered)
````

이 코드는 간단한 예제 데이터에서 IQR 방식을 사용하여 이상치를 탐지하고 해당 이상치를 포함하는 행을 제거하는 방법을 보여줍니다.

대부분의 머신러닝 알고리즘은 <b>지도학습(supervised learning)</b> 방식을 사용합니다. 이 방식에서는 <b>X</b>에 해당하는 독립변인과 <b>y</b>에 해당하는 종속변인이 함께 제공됩니다. 여기서 <b>X</b>는 특징(feature) 또는 입력(input) 데이터, <b>y</b>는 목표(target) 또는 라벨(label) 데이터라고도 불립니다. 알고리즘은 이 <b>X</b>와 <b>y</b>의 관계를 학습하여 새로운 입력 데이터에 대한 예측을 수행합니다. 예를 들어, 집의 특징(면적, 방 개수 등)인 <b>X</b>를 사용하여 집의 가격인 <b>y</b>를 예측하는 것입니다.<br><br>

결론적으로, 데이터를 <b>X</b>와 <b>y</b>로 나누는 것은 머신러닝 모델을 학습시키고 평가하는 과정에서 필수적입니다.

본 프로젝트 데이터인 'train' 데이터를 예로 든 설명입니다.

1. __데이터셋 설정__ : 학습 및 검증에 사용할 데이터(train)을 지정합니다.
2. __데이터 분할__ : 'credit' 열을 타겟 변수(y)로, 나머지 모든 열을 특성(X)로 설정합니다.
3. __train_test_split 사용__ : train_test_split 함수를 사용하여 데이터를 학습 세트와 테스트 세트로 일정 비율로 분할합니다. 이 함수는 무작위로 데이터를 섞은 후 분할하므로, 동일한 결과를 얻기 위해 random_state 매개변수를 설정합니다.

>__train_test_split의 주요 매개변수:__
>
>- arrays: 분할하려는 데이터 (예: X, y)
>- test_size: 테스트 세트의 비율 또는 크기 (0과 1 사이의 실수로 표시)
>- train_size: 학습 세트의 비율 또는 크기. test_size가 지정되어 있을 경우 생략 가능
>- random_state: 재현성을 위한 시드 값
>- shuffle: 데이터를 분할하기 전에 데이터를 섞을 것인지 결정 (기본값은 True)

## 파생변수

### 파생변수 생성
- 파생변수 생성의 이유:
1. <b>정보 가치 향상</b>: 원본 데이터셋에 없는 새로운 정보나 패턴을 포착하기 위해 파생변수를 만듭니다.
2. <b>데이터의 해석력 향상</b>: 비즈니스 문제나 도메인 지식을 기반으로 파생변수를 만들어 데이터의 해석력을 높일 수 있습니다.
3. <b>모델 성능 향상</b>: 더 많은 정보를 제공하여 모델의 예측 성능을 향상시킬 수 있습니다.
4. <b>데이터의 차원 축소</b>: 여러 변수의 정보를 합쳐 하나의 파생변수로 축소할 수 있습니다.

#### 예시
아래 코드는 간단한 예제 데이터를 사용하여 파생변수를 생성하는 방법을 보여줍니다. 가정은 연간 소득과 나이를 기반으로 한 '소득 대비 나이'라는 파생변수를 생성하는 것입니다.

````python
# 예제 데이터 생성
data = {
    'Age': [25, 30, 35, 40, 45],
    'Annual_Income': [50000, 55000, 60000, 65000, 70000]
}
df = pd.DataFrame(data)

# 파생변수 생성: 소득 대비 나이
df['Income_per_Age'] = df['Annual_Income'] / df['Age']

print(df)
````

위의 코드에서는 각 사람의 연간 소득을 그 사람의 나이로 나누어 '소득 대비 나이'라는 새로운 파생변수를 생성합니다. 이렇게 생성된 파생변수는 특정 연령대의 사람들이 얼마나 수입을 얻는지에 대한 정보를 제공할 수 있습니다.

실제 비즈니스 문제나 데이터셋에 따라 도메인 지식을 활용하여 다양한 파생변수를 생성할 수 있습니다.

이제 실습데이터에 대해 파생변수를 생성하고 그에 대해 설명하겠습니다.

````python
import numpy as np # 전처리 라이브러리 임포트
# 파생변수 생성 및 주석

# 1. 파생변수 생성을 위해 음수값 양수로 전환
# 일부 변수들에서의 일수는 음수로 표시되어 있어 계산의 편의성을 위해 양수로 전환합니다.
negative_list = ['DAYS_EMPLOYED', 'DAYS_BIRTH', 'begin_month']
for i in negative_list:
    train[f"{i}"] = -train[f"{i}"]

# 2. 사용자의 나이 계산
# 사용자가 태어난 지 몇 일이 지났는지를 기반으로 나이를 계산합니다.
train['Age'] = abs((train['DAYS_BIRTH']) / 365)

# 3. 사용자의 고용된 연수 계산
# 얼마나 오래 전에 고용되었는지를 나타내는 변수입니다.
train['Years_Employed'] = train['DAYS_EMPLOYED'].apply(lambda x: -x / 365 if x < 0 else 0)  # 양수값은 무직자로 간주하고 0 처리
train.loc[train['DAYS_EMPLOYED'] < 0, 'DAYS_EMPLOYED'] = 0

# 4. 사용자의 연락 수단 수 계산
# 사용자가 가지고 있는 연락 수단의 총 수를 계산합니다.
train['Contact_Count'] = train['FLAG_MOBIL'] + train['work_phone'] + train['phone'] + train['email']

# 5. 소득 대비 자녀 수
# 각 사용자의 소득에 대한 자녀 수의 비율을 계산합니다.
train['income_per_child'] = train['income_total'] / (train['child_num'] + 1)

# 6. 부양가족 수당 소득
# 사용자의 총 소득을 부양가족 수로 나누어 한 부양가족 당 평균 소득을 계산합니다.
train['income_per_family'] = train['income_total'] / (train['family_size'])

# 7. 고용전 일수
# 사용자가 처음으로 고용되기 전까지의 전체 일수를 계산합니다.
train['before_EMPLOYED'] = train['DAYS_BIRTH'] - train['DAYS_EMPLOYED']

# 8. 고용되기 전까지의 시간 대비 총 소득
# 사용자의 총 소득을 고용되기 전까지의 일수로 나눈 값을 계산합니다.
train['income_total_befofeEMP_ratio'] = train['income_total'] / train['before_EMPLOYED']

# 9. 사용자가 처음으로 고용되기 전까지의 월 수
# 사용자가 처음으로 고용되기 전까지의 월 수를 계산합니다.
train['before_EMPLOYED_m'] = np.floor(train['before_EMPLOYED'] / 30) - ((np.floor(train['before_EMPLOYED'] / 30) / 12).astype(int) * 12)

# 10. 가장의 명수
# 부양 가족 수에서 자녀의 수를 뺀 값은 경제 활동을 할 수 있는 가장의 수를 나타냅니다.
train["active_parents"] = train["family_size"]-train["child_num"]
train.loc[train['active_parents'] < 0, 'active_parents'] = 0

# 출력
train.describe()
````

#### 설명
1. __음수값을 양수로 전환__: 일부 변수들에서의 일수는 음수로 표시되어 있습니다. 이러한 음수값은 데이터셋에서 어떤 기준 시점으로부터의 경과일을 나타내기 때문입니다. 이 값들을 양수로 변환하여 계산의 편의성을 높입니다.
2. __Age__: DAYS_BIRTH 변수를 사용하여 사용자의 나이를 계산합니다. DAYS_BIRTH는 사용자가 태어난 지 몇 일이 지났는지를 나타내므로, 이를 365로 나누어 나이를 계산합니다.
3. __Years_Employed__: 사용자가 얼마나 오래 전에 고용되었는지를 나타내는 변수입니다. DAYS_EMPLOYED의 값은 고용된 일수를 나타냅니다. 양수 값은 무직 상태를 나타내므로 0으로 처리하고, 음수 값은 고용된 기간을 나타냅니다.
4. __Contact_Count__: 사용자가 가지고 있는 연락 수단의 총 수를 나타냅니다. 여기에서는 모바일, 직장 전화, 개인 전화, 이메일의 유무를 합하여 총 연락 수단 수를 계산합니다.
5. __income_per_child__: 이 파생변수는 각 사용자의 소득에 대한 자녀 수의 비율을 나타냅니다. 소득이 높아도 자녀 수가 많으면 부담이 클 수 있으므로, 이 비율로 그 부담을 나타낼 수 있습니다.
6. __income_per_family__: 사용자의 총 소득을 부양가족 수로 나누어 한 부양가족 당 평균 소득을 계산합니다.
7. __before_EMPLOYED__: before_EMPLOYED는 사용자가 처음으로 고용되기 전까지의 전체 일수를 나타냅니다. 이는 사용자의 전체 생애 (DAYS_BIRTH)에서 고용된 일수 (DAYS_EMPLOYED)를 뺀 값으로 계산됩니다.
8. __income_total_befofeEMP_ratio__: 사용자의 총 소득을 고용되기 전까지의 일수로 나눈 값을 나타냅니다.
9. __before_EMPLOYED_m__: 사용자가 처음으로 고용되기 전까지의 월 수를 나타냅니다.
10. __active_parents__: 부양 가족 수에서 자녀의 수를 뺀 값입니다. 이 값은 경제 활동을 할 수 있는 부모님의 수를 나타냅니다.<br>

### Kmeans 클러스터링 학습
KMeans 클러스터링은 비지도 학습 알고리즘 중 하나로, 주어진 데이터를 'K'개의 클러스터로 그룹화하는 알고리즘입니다. 여기서 'K'는 사용자가 지정해야 하는 클러스터 수를 나타냅니다.

#### KMeans의 작동 원리

1. 초기 'K'개의 클러스터 중심을 데이터 포인트에서 무작위로 선택하거나 다른 방법으로 초기화합니다.
2. 각 데이터 포인트를 가장 가까운 클러스터 중심에 할당합니다.
3. 각 클러스터의 중심을 해당 클러스터의 데이터 포인트 평균으로 업데이트합니다.
4. 클러스터 중심의 이동이 없을 때까지 2번과 3번 단계를 반복합니다.<br>

#### KMeans의 주요 특징
+ 초기 중심값의 선택: KMeans의 결과는 초기 중심값의 선택에 따라 달라질 수 있으므로, 여러 번 실행하거나 초기화 방법을 변경해보는 것이 좋습니다.
+ 거리 측정: 통상적으로 유클리디안 거리를 사용하여 데이터 포인트와 클러스터 중심 간의 거리를 측정합니다.
+ 클러스터 수 'K'의 선택: 'K'값은 사용자가 지정해야 하는데, 이를 선택하기 위한 방법으로 Elbow 방법 등이 있습니다.

 ##### KMeans 알고리즘 적용 및 WCSS 값 계산
이제 변환된 수치형 데이터를 사용하여 KMeans 알고리즘을 적용하고, 다양한 k 값에 대한 WCSS 값을 계산하겠습니다. WCSS는 클러스터 내의 제곱 거리의 합을 나타내며, 이 값을 사용하여 최적의 k 값을 선택하는 데 도움을 받을 수 있습니다.

WCSS 값이 급격히 감소하는 지점, 즉 "elbow point"를 찾아서 최적의 클러스터 개수를 결정할 수 있습니다. 여기서는 k의 범위를 1부터 10까지로 설정하고 WCSS 값을 계산해보겠습니다.

````python
# 변환된 수치형 데이터 추출
kmean_target = kmeans_data[['income_type', 'edu_type']]

# 다양한 k 값에 대한 WCSS 값을 저장할 리스트 초기화
wcss = []

# k = 1부터 10까지 KMeans 알고리즘 적용 및 WCSS 값 계산
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(kmean_target)
    wcss.append(kmeans.inertia_)

wcss
````

WCSS 값이 계산되었습니다.

모델의 성능은 불필요한 변수를 제거함으로써 향상될 수 있으며 여러 가지 이유가 있습니다:<br>

__1. 차원의 저주 (Curse of Dimensionality):__

고차원의 데이터 공간에서는 데이터 포인트 간의 거리가 크게 증가하기 때문에, 예측 모델이 올바른 판단을 내리기가 어려워집니다. 이는 특히 군집화, 최근접 이웃 검색과 같은 거리 기반 알고리즘에서 중요합니다.
불필요한 변수를 제거함으로써 차원을 줄이면, 데이터 포인트 간의 관계를 더 쉽게 파악할 수 있게 됩니다.<br>
__2. 과적합 (Overfitting):__

불필요한 변수가 많을수록 모델은 훈련 데이터에 대한 불필요한 패턴까지 학습할 위험이 있습니다. 이로 인해 테스트 데이터나 실제 환경에서의 성능이 저하될 수 있습니다.
변수를 제거하면 모델의 복잡성이 감소하고, 과적합을 방지할 수 있습니다.<br>
__3. 계산 효율성:__

변수의 수가 줄어들면 학습, 예측, 검증 과정의 계산 비용이 줄어듭니다. 이로 인해 모델의 학습과 추론 속도가 빨라집니다.<br>
__4. 모델의 해석성 (Interpretability):__

중요하지 않은 변수가 제거되면, 남아있는 변수들 사이의 관계가 더 명확해집니다. 따라서 모델의 예측을 해석하기가 더 쉬워집니다.<br>
__5. 다중공선성 (Multicollinearity):__

변수들 사이에 강한 상관관계가 있을 경우, 다중공선성이 발생할 수 있습니다. 이는 회귀 모델에서 계수의 추정치가 불안정해질 수 있게 만듭니다.
불필요한 변수를 제거함으로써 다중공선성 문제를 줄일 수 있습니다.<br>
__6. 노이즈 감소:__

불필요한 변수는 종종 데이터의 노이즈를 증가시키는 원인이 됩니다. 이러한 변수를 제거함으로써 노이즈가 감소하고, 모델의 일반화 성능이 향상될 수 있습니다.<br>
__7. 특성 선택의 중요성:__

특성 선택을 통해 가장 중요한 변수들만을 사용하면, 해당 변수들이 예측에 얼마나 큰 영향을 미치는지 더 잘 이해할 수 있습니다.
요약하면, 불필요한 변수를 제거함으로써 모델의 성능은 물론, 학습 속도, 모델의 해석성 등 여러 면에서 이점을 얻을 수 있습니다.

이제 모델의 Feature Importances를 추출하고 그에 맞게 변수를 조절하며 모덜 성능을 개선해보겠습니다.

````python
import matplotlib.pyplot as plt
importances = xgb_best.feature_importances_

# Feature names 가져오기
features = X_train.columns

# Feature importance와 함께 정렬된 순서의 index 가져오기
indices = np.argsort(importances)

# Plot
plt.figure(figsize=(10, 12))
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
````

그래프에 따르면 'Years_Employed', 'reality_Y', 'FLAG_MOBIL' 등과 같이 모델 분석에 영향력이 낮은 변수들을 확인할 수 있습니다.

이제 낮은 중요도의 변수를 제거하고 모델을 재학습시키겠습니다.

````python
# Feature importances를 낮은 중요도 순으로 정렬
sorted_indices = np.argsort(importances)

# 정렬된 순서에 따라 feature names 가져오기
sorted_features = [features[i] for i in sorted_indices]

print(sorted_features)
````

````python
# drop 변수 리스트
drop_list = sorted_features[:6]

# 모델 학습
xgb_best.fit(X_train.drop(columns = drop_list, axis = 1), y_train.drop(columns = drop_list, axis = 1))

# 추론
best_pred = xgb_best.predict(X_val.drop(columns = drop_list, axis = 1))

# 성능 평가
accuracy_2 = accuracy_score(y_val, best_pred)
precision_2 = precision_score(y_val, best_pred, average='macro')
recall_2 = recall_score(y_val, best_pred, average='macro')
f1_2 = f1_score(y_val, best_pred, average='macro')

print("Accuracy: ", accuracy_2)
print("Precision: ", precision_2)
print("Recall: ", recall_2)
print("F1-score: ", f1_2)
````
